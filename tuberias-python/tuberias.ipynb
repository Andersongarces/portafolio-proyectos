```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("tuberias-python.xlsx")
df.head()
```

```python
colors = {'enero': 'red', 'febrero': 'green', 'marzo': 'blue'}

plt.figure(figsize=(10, 7))

for month, color in colors.items():
    subset = df[df['Mes'] == month]
    plt.scatter(subset['Longitud por pie'], subset['Total de solicitudes por servicio'], label=month, color=color, s=subset['Total de solicitudes por servicio']*2, alpha=0.7)

plt.xlabel('Longitud por pie')
plt.ylabel('Total de solicitudes por servicio')
plt.title('Total de fugas en comparacin con la longitud de las tuberias')

# Aadir leyenda con burbujas del mismo tamao
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=8, label=month) for month, color in colors.items()]
plt.legend(handles=legend_handles)

plt.show()
```

```python
import matplotlib.pyplot as plt

colors = {'enero': 'red', 'febrero': 'green', 'marzo': 'blue'}

# Ajusta el tamao de la figura en funcin de los datos
fig, ax = plt.subplots(figsize=(len(colors) * 5, 7))

for month, color in colors.items():
    subset = df[df['Mes'] == month]
    # Utiliza la escala dinmica basada en los datos
    plt.scatter(subset['Longitud por pie'], subset['Total de solicitudes por servicio'], label=month, color=color, s=subset['Total de solicitudes por servicio']*2, alpha=0.7)

plt.xlabel('Longitud por pie')
plt.ylabel('Total de solicitudes por servicio')
plt.title('Total de fugas en comparacin con la longitud de las tuberas')

# Aade leyenda con burbujas del mismo tamao
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=8, label=month) for month, color in colors.items()]
plt.legend(handles=legend_handles)

plt.show()
```
