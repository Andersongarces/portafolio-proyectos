> #### Previo a responder las preguntas es necesario realizar un anlisis exploratorio y una limpieza de datos



```python
df = pd.read_csv("mission-launches.csv")
df.head(5)
```

```python
df.info()
```

```python
porcentajes_nulos = df.isnull().sum()*100/df.shape[0]
porcentajes_nulos = round(porcentajes_nulos, 2)
porcentajes_nulos
```

```python
df["Price"].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

---

#### Con el anlisis anterior ya no hay valores nulos en la columna Price
#### Sin embargo en la columna Location hay ms de un registro. Esto no permite obtener la localizacin exacta
#### En la columna Rocket_Status el registro est unido al 'estado' por lo que hay que separarlo
#### Y en la columna Date no est en meses por lo tanto no se podra hacer un grfico 

> #### Por lo tanto el archivo amerita limpieza de datos

```python
df['Country'] = df["Location"].str.split(',').str[-1].str.strip() 
df['Country'].sample(10)
```

```python
df["Date"].head(5)
```

```python
df["DateTime"] = df["Date"].str[:16]
df["DateTime"] = pd.to_datetime(df["DateTime"])
df["DateTime"]
```

```python
df["Year"] = pd.to_datetime(df["DateTime"])
df["Year"] = df["Year"].dt.year
```

```python
df.head(5)
```

```python
columnas = []
```

```python
df_clean1 = df.drop(df.columns[:2], axis=1)
df_clean1.info()
```

```python
df_clean2 = df_clean1.drop(df_clean1.columns[1:4], axis=1)
df_clean2.info()
```

```python
df_clean3 = df_clean2.drop(df_clean2.columns[-2], axis=1)
df_clean3.info()
```

```python
df_clean4 = df_clean3.drop(df_clean3.columns[1], axis=1)
df_clean4.info()
```

```python
df_clean4["Price"] = pd.to_numeric(df_clean4["Price"], errors='coerce')

df_clean4["Price"] = df_clean4["Price"].fillna(0).astype(int)

df_clean4["Price"]
```

```python
df_clean4.info()
```
